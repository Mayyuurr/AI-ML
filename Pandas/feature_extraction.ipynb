```python
def extract_eps(title):
    check=False
    eps=''
    for i in title:
        if i==')':
            check=False
            break
        if check==True:
            eps+=i
        if i=='(':
            check=True
    return eps
```

```python
df['Episodes']=df['Title'].apply(extract_eps)
df.head()
```

```python
df['Episodes']=df['Episodes'].str.replace(' eps','')
df['Episodes']=df['Episodes'].astype(int)
df.head()
```

```python
def extract_time(title):
    time=''
    for i in range(len(title)):
        if title[i]==')':
            for j in range(i+1,i+20):
                time+=title[j]
    return time
```

```python
df['time Stamp']=df['Title'].apply(extract_time)
```

```python
from datetime import datetime

def get_total_months(date_str):

    # 1. Split the string into Start and End
    start_part, end_part = date_str.split(' - ')
    
    # 2. Convert strings to datetime objects (%b is for short month names like 'Apr')
    start_date = datetime.strptime(start_part.strip(), '%b %Y')
    end_date = datetime.strptime(end_part.strip(), '%b %Y')
    
    # 3. Calculate the difference
    years_diff = end_date.year - start_date.year
    months_diff = end_date.month - start_date.month
    
    # Add 1 if you want the count to be inclusive (e.g., Jan to Jan = 1 month)
    total_months = (years_diff * 12) + months_diff + 1
    return total_months

```

```python
df['Months'] = df['time Stamp'].apply(get_total_months)
df.head()
```
