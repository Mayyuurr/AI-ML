```python
import numpy as np
import pandas as pd
```

#### **Finding missing data**

```python
data={
    'A':[1,2,np.nan,4,5],
    'B':[np.nan,2,3,4,5],
    'C':[1,2,3,np.nan,np.nan],
    'D':[1,np.nan,np.nan,np.nan,5]
}

df=pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
df.isna().sum()
#count of null values
```

```python
df.isna().any()
#check if any null value present in each (deafualt axis=0) can chaneg axis=0 to get any null value in row(google this not sure)
```

#### **Removing missing data**

```python
df
```

```python
df.dropna()
#as there are null value in all rows thus every row dropped
```

```python
df.at[1,'D']=2
df
```

```python
df.dropna(thresh=3)
# drop rows having less tha 3 null values
```

#### **Filling the missing data**

```python
df
```

```python
df.fillna(0)
```

```python
value_list={'A':0,'B':10,'C':100,'D':1000}
df.fillna(value=value_list)
```

```python
df.fillna(df.mean())
```
