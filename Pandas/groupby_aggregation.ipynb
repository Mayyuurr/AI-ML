```python
import numpy as np
import pandas as pd
```

#### **GroupBy**

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2026-01-01', periods=8)
}
df=pd.DataFrame(data)
df
```

```python
category_group=df.groupby('Category')
for A, B in category_group:
    print(A)
    print(B)
```

```python
#Group by category and calculate sum of sales
#df.groupby('Category')['Sales'].sum()
category_group['Sales'].sum()
```

```python
#Group by store and calculate sum of sales
store_group=df.groupby('Store')
store_group['Sales'].sum()
```

```python
#group by multiple columns and calcluate sales
#Group by Category and Store
cat_store_grp=df.groupby(['Category','Store'])
for cat, store in cat_store_grp:
    print(cat)
    print(store)
```

```python
cat_store_grp['Sales'].sum()

```
