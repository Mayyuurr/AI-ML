```python
import numpy as np
import pandas as pd
```

#### **Pivot Tables**

a Pivot Table is a tool used to summarize, sort, reorganize, group, count, total, or average data stored in a table. It allows you to transform "long" data into a "wide" format to see relationships between variables clearly.

```python
data = {
    'Date': pd.date_range('2026-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}

df=pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
df
```

```python
pivot1=pd.pivot_table(df,values='Sales',index='Region',columns='Product',aggfunc='median')
```

```python
pivot2=pd.pivot_table(df,values=['Sales','Units'],index='Region' ,columns='Product',aggfunc="median")
pivot2
```

#### **Cross Tabs**

a Crosstab (short for Cross Tabulation) is a specific type of pivot table used to show the frequency distribution of variables. While a pivot table can calculate sums or averages, a crosstab is primarily used to count how many times combinations of categories occur.

```python
pd.crosstab(df['Region'],df['Product'])
```
