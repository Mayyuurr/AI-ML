```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv(r'Countries.csv')
df.head()
```

```python

#which contry has highest population

# highest_population_cnty=df['population'].idxmax()
# print(df.loc[highest_population_cnty])

df[df['population']==df['population'].max()]
```

```python
#what is the capital of country with highest population

highest_population_cnty=df['population'].idxmax()

df.loc[highest_population_cnty]['capital_city']
```

```python
#which country has the least population 
df[df['population']==df['population'].min()]
```

```python
    #what is the capital of country with least population
df.loc[df['population'].idxmin()]['capital_city']
```

```python
#give top 5 countries with highest democratic score
df.nlargest(5,'democracy_score')
```

```python
#how many total regions are there
df['region'].count #Gives Each region with count from 1
df['region'].count()  #Gives number of total regions count from 0
```

```python
#list of unique regions

df['region'].unique()
```

```python
#how many countries lie in Eastern Europe region
eastern_europe=df[df['region']=='Eastern Europe']
eastern_europe['country'].count()
eastern_europe['country']
```

```python
#who is political leader of the 2nd highest populated country
most_populated_2=df.nlargest(2,'population')
most_populated_2.iloc[1]['political_leader']
```

```python
#how many countries are there whose political leader is unknown 
unknown_leader=df[df['political_leader'].isnull()]
unknown_leader['country']
```

```python
#how many contry have republic in their full name
count=0
def republic_check(country):
    global count
    if 'republic' in country.lower():
        count+=1
    return country

df['country_long']=df['country_long'].apply(republic_check)
print(count)
```

```python
# which contry in african region has has highest population

africa_continent=df[df['continent']=='Africa']
highest_pop_row=africa_continent[africa_continent['population']==africa_continent['population'].max()]
highest_pop_row['country']
```
