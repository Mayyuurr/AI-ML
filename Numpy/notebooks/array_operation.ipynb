#### Brodacasting

Broadcasting is one of the most powerful features in NumPy. It allows you to perform arithmetic operations (like addition or multiplication) on arrays of different shapes without having to manually duplicate data to make the sizes match

```python
#[10,20,30,40,50]
array_brodcasting=np.arange(10,51,10)
array_brodcasting
```

```python
array_brodcasting+10
```

```python
l=[10,20,30,40]
list_arr=np.array(l)
list_arr
```

```python
list_arr+10
```

#### Deep and Shallow copy

```python
array_org=np.arange(1,21)
array_org
```

```python
array_org[:5]*10
#shallow copy does not change original array
```

```python
array_org
```

```python
array_copy=array_org
array_copy[0]=99
array_copy
#deep copy i.e original array changed 
#both point to same memory location so change by one affect other meanwhile in shallow it point to completely different memory locations
```

```python
array_org
```

### Advance Array Manipulation

#### Stacking Array

```python
a=np.array([1,2,3,4,5])
b=np.array([6,7,8,9,10])
```

```python
np.hstack((a,b))
```

```python
np.vstack((a,b))
```

```python
np.column_stack((a,b))
```

```python
np.row_stack((a,b))
```

#### Splitting Array

```python
c=np.arange(16).reshape(4,4)
c
```

```python
np.split(c,2,1)
#split (array_name,divide_in_parts,axis)
```

```python
np.hsplit(c,2)
```

```python
split_array=np.vsplit(c,2)
split_array
```
