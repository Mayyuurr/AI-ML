```python
import numpy as np
```

```python
arr=np.arange(11,21)
arr
```

### Array/Vector Indexing and Slicing

```python
arr[2:5]
```

```python
arr[1:9:2]
```

### Numpy Indexing and Slicing of Matrix

```python
arr=np.arange(1,31).reshape(6,5)
arr
```

```python
arr[3,4]
#[row_index,col_index]
```

```python
slice=arr[0:2,1:4]
slice
```

```python
slice_block=arr[3:,3:]
slice_block
```

```python
slice_row=arr[1]
slice_row
```

```python
slice_column=arr[:,1]
slice_column
```

### Boolean Indexing

```python
arra=np.arange(11,21)
arra
```

```python
bool_index=arra%2==0
bool_index
```

```python
arra=arra[bool_index]
arra
```
