```python
import seaborn as sns
import matplotlib.pyplot as plt
```

We use distribution plot to visualize the distribution of quantitative data.

Lets discuss some plots that allo us to visualize the distribution of data set. These plots are:
- distplot/histplot
- jointplot
-  pairplot
- rugplot

```python
df=sns.load_dataset('tips')
df
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde=True)
```

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind='scatter')
```

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind='reg')
```

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind='kde')
```

```python
sns.pairplot(data=df)
```

```python
sns.pairplot(data=df,hue='size',palette='rainbow')
```

```python
sns.rugplot(df)
```
