```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### **MatPlotlib  basics**

```python
array1=np.linspace(0,5,11)
array1
```

```python
array2=array1**2
array2
```

```python
plt.plot(array1,array2)
plt.title("New Plot")
plt.xlabel('Array1')
plt.ylabel('Array2')
```

#### **Subplots in matplotlib**

```python
plt.subplot(2,2,1)
plt.plot(array1,array2)
plt.subplot(2,2,3)
plt.plot(array2,array1)
plt.subplot(2,2,4)
plt.plot(array1,array1)
plt.subplot(2,2,2)
plt.plot(array2,array1**2)
```

#### **Object Oriented way**

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
fig=plt.figure(figsize=(15,10),dpi=100)

#add_axes(rect, projection=None, polar=False, **kwargs)
#rect : tuple (left, bottom, width, height)

axis1=fig.add_axes([0.1,0.1,0.5,0.5])
axis1.plot(array1,array2)
axis2=fig.add_axes([0.7,0.1,0.5,0.5])
axis2.plot(array2,array1)
```

#### **Types of Plot**

```python
plt.scatter(array1,array2)
```

```python
plt.hist(array1,array2)
```

```python
plt.boxplot(array2)
```

#### **Saving plots**

```python
save_fig=plt.figure()
ax=save_fig.add_axes([0.1,0.1,0.5,0.5])
ax.plot(array1,array2)
plt.savefig('Basic_Fig_save.png')
```
